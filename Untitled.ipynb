```python
import os
import numpy as np
import cv2
import tensorflow as tf
import pandas as pd
```

```python
directory_im = "C:\\Users\\yashl\\Documents\\Sample Train\\Image\\2019_04_09_cms1000"
directory_lb = "C:\\Users\\yashl\\Documents\\Sample Train\\Label\\2019_04_09_cms1000"
```

```python
image_paths = []
for root, directories, files in os.walk(directory_im):
    for filename in files:
        file_path = os.path.join(root, filename)
        image_paths.append(file_path)

```

```python
label_paths = []
for root, directories, files in os.walk(directory_lb):
    for filename in files:
        file_path = os.path.join(root, filename)
        label_paths.append(file_path)
```

```python
#checking for empty CSVs
for label_path in label_paths:
    with open(label_path, 'r') as f:
            lines = f.readlines()
        # check if the file is empty
    if len(lines) == 0:
            print(f'{label_path} is empty')
```

```python
#jysyrmsslxbjrngbli
X_train = []
for img_path in image_paths:
    image = cv2.imread(img_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_resized = tf.image.resize(image, (224,224))
    X_train.append(image_resized.numpy())
X_train = np.array(X_train, dtype = np.uint8)
```
